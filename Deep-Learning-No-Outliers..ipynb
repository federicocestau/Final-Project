```python
Financial_Data.drop(columns=['Unnamed: 0'],inplace=True)
```

```python
def make_lags(df,n,col):
    df[f'{col}_lagged_{n}']=df.groupby('Ticker')[col].shift(n)
    #df = df.dropna()
    return df
```

```python
# call the function to create 4 new lagged columns in a dataframe called `df`
#for i in range(1,5):
    #df=make_lags(Financial_Data,i,'EPS')
#Using data from 4 time period before the data we're trying to predict. 
for col in ['EPS-Annual Basis', 'CFPS', 'PE_Ratio', 'Market_Cap',
       'ROE', 'Price_Sales_Ratio', 'Debt_Equity_Ratio']:
    df = make_lags(Financial_Data, 2, col)
df = df.dropna()
```

```python
df["Quarter-Year"].value_counts()
```

```python
df
```

```python
#We split the data (208+53=261).208/261=79,69%
#79.69% it is my train data all the quarters before 2018
#20.31% it is my test data which is the second quarter 2018, what we are trying to predict
Train = df[(df['Quarter-Year']<2018)]
Test = df[(df['Quarter-Year']>=2018)]
print(Train.shape)
print(Test.shape)
```

```python
Y_Train = Train['Final_Rating']
Y_Test = Test['Final_Rating']
```

```python
Train.columns
```

```python
X_Train = Train[['EPS-Annual Basis_lagged_2',
       'PE_Ratio_lagged_2', 'Market_Cap_lagged_2', 'ROE_lagged_2',
       'Price_Sales_Ratio_lagged_2',
       'Debt_Equity_Ratio_lagged_2']]
X_Test = Test[['EPS-Annual Basis_lagged_2',
       'PE_Ratio_lagged_2', 'Market_Cap_lagged_2', 'ROE_lagged_2',
       'Price_Sales_Ratio_lagged_2',
       'Debt_Equity_Ratio_lagged_2']]
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_Train)
```

```python
X_train_scaled = X_scaler.transform(X_Train)
X_test_scaled = X_scaler.transform(X_Test)
```

```python
print(Train.columns)
```

```python
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
# One-hot encoding
y_train_categorical = to_categorical(Y_Train)
y_test_categorical = to_categorical(Y_Test)
y_train_categorical
```

```python
#first, create a normal neural network with 2 inputs, 6 hidden nodes, and 2 outputs
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=6))
model.add(Dense(units=6, activation='relu'))
#model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
# Compile the model
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Fit the model to the training data
model.fit(
    X_train_scaled,
    y_train_categorical,
    epochs=100,
    shuffle=True,
    verbose=2
)
```

```python
model_loss, model_accuracy = model.evaluate(X_train_scaled, y_train_categorical, verbose=2)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Normal Neural Network - Loss: {model_loss}, Accuracy: {model_accuracy}")
```
