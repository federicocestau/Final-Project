```python
base_url3 = "https://financialmodelingprep.com/api/v3/key-metrics/"
query3 = "?apikey=12ca4a34c2a88857222c5812d1fcfc50"
```

```python
ticker_list = ["AMRS", "CGC","TLRY","CARA","CRBP","IIPR","NBEV","TRST","YCBD","CRON","GWPH",
              "SMG","TAP","MO","BUD",]
```

```python
Net_Income_Per_Share=[]
Anual_Dates = []
Ticker = []
for ticker in ticker_list:
    print("Try: " + ticker)

  
    url = base_url3 + ticker + query3
    response = requests.get(url).json()
    
    for d in response:
        Net_Income_Per_Share.append(d['netIncomePerShare'])
        Anual_Dates.append(d['date'])
        Ticker.append(d['symbol'])
```

```python
Financial_DF_EPS = pd.DataFrame({
    "F_Date_EPS": Anual_Dates,
    "EPS": Net_Income_Per_Share,
    "Ticker":Ticker
    })
```
