```python
#Dependencies
import json
import requests
API_KEY= ENTER API KEY HERE
import pandas as pd
```

```python
base_url = "https://financialmodelingprep.com/api/v3/key-metrics/"
query = f"?period=quarter&apikey={API_KEY}"
```

```python
ticker_list = ["AMRS", "CGC","TLRY","CARA","CRBP","IIPR","NBEV","TRST","YCBD","CRON","GWPH",
              "SMG","TAP","MO","BUD",]
```

```python
Ticker = []
Date=[]
Net_Income_Per_Share=[]
Free_Cash_Flow_Per_Share=[]
PE_Ratio=[]
Market_Cap=[]

for ticker in ticker_list:
    
    url = base_url + ticker + query
    response = requests.get(url).json()
    # print(response)
    for d in response:
        print(d['symbol'] + " " + d['date'])

        Date.append(d['date'])
        Net_Income_Per_Share.append(d['netIncomePerShare'])
        Free_Cash_Flow_Per_Share.append(d['freeCashFlowPerShare'])
        PE_Ratio.append(d['peRatio'])
        Market_Cap.append(d['marketCap'])
        Ticker.append(d['symbol'])
```

```python
base_url2= "https://financialmodelingprep.com/api/v3/ratios/"
query2 = f"?period=quarter&apikey={API_KEY}"
```

```python
ROE = []
Price_Sales_Ratio = []
Payout_Ratio = []
Debt_Equity_Ratio = []
Dividend_Yield = []

for ticker in ticker_list:
    print("Try: " + ticker)

  
    url2 = base_url2 + ticker + query2
    response = requests.get(url2).json()

    for d in response:
        Price_Sales_Ratio.append(d['priceToSalesRatio'])
        Payout_Ratio.append(d['payoutRatio'])
        ROE.append(d['returnOnEquity'])
        Debt_Equity_Ratio.append(d['debtEquityRatio'])
        Dividend_Yield.append(d['dividendYield'])
        
        
```
