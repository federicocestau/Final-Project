```python
#A lag is taking previous data and testing it with present data. 
def make_lags(df,n,col):
    df[f'{col}_lagged_{n}']=df.groupby('Ticker')[col].shift(n)
    #df = df.dropna()
    return df
```

```python
# call the function to create 2 new lagged to predict data two periods from today,2 quarters.
for col in ['EPS-Annual Basis', 'CFPS', 'PE_Ratio', 'Market_Cap',
       'ROE', 'Price_Sales_Ratio', 'Debt_Equity_Ratio']:
    df = make_lags(Financial_Data, 2, col)
df = df.dropna()
```

```python
df["Quarter-Year"].value_counts()
```

```python
df
```

```python
Train = df[(df['Quarter-Year']<2018)]
Test = df[(df['Quarter-Year']>=2018)]
print(Train.shape)
print(Test.shape)
```

```python
#Final Rating is my dependent variable. If it is 0 means that EPS <0, 
#if it is 1 means that EPS >=0
Y_Train = Train['Final_Rating']
Y_Test = Test['Final_Rating']
```

```python
#Train Data is the one we use to create (Fit into) the model. Test data is the one we use to test how good
#is our model measure by its final score. 
Train.columns
```

```python
#These are the variables we will use to fit into our model
X_Train = Train[['EPS-Annual Basis_lagged_2',
       'PE_Ratio_lagged_2', 'Market_Cap_lagged_2', 'ROE_lagged_2',
       'Price_Sales_Ratio_lagged_2',
       'Debt_Equity_Ratio_lagged_2']]
#These are the same variables we will use to test our model
X_Test = Test[['EPS-Annual Basis_lagged_2',
       'PE_Ratio_lagged_2', 'Market_Cap_lagged_2', 'ROE_lagged_2',
       'Price_Sales_Ratio_lagged_2',
       'Debt_Equity_Ratio_lagged_2']]
```

```python
#Scaling the data
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_Train)
```

```python
X_train_scaled = X_scaler.transform(X_Train)
X_test_scaled = X_scaler.transform(X_Test)
```

```python
print(Train.columns)
```

```python
#We split the data (208+53=261).208/261=79,69%
#79.69% it is my train data all the quarters before 2018
#20.31% it is my test data which is the second quarter 2018, what we are trying to predict
print(X_Train.shape)
print(X_Test.shape)
print(Y_Train.shape)
print(Y_Test.shape)
```

```python
#create the model and fit the data
from xgboost import XGBRegressor
classifier=XGBRegressor()
```

```python
#Fitting the train data.
classifier.fit(X_train_scaled, Y_Train)
```

```python
print(f"Training Data Score: {classifier.score(X_train_scaled, Y_Train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, Y_Test)}")
```
